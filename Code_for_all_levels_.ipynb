```
# Code Lexicon: Data Logic and Retention Handbook
# This lexicon illustrates the three core code logics within this system
# and demonstrates how data ('Hello World') is processed and retained at each layer.

# --- 1. The Base Layer: Immortal Binary Assertion ---
# Purpose: Represents data as a fundamental, immutable assertion of existence.
# Data Retention: The original message is encoded into a sequence of binary values,
# ensuring its core representation is preserved in its most atomic form.

def immortal_binary_hello():
    # Each segment corresponds to a localized state of existence
    code = [0b01101000, 0b01100101, 0b01101100, 0b01101100, 0b01101111,
            0b00100000, 0b01110111, 0b01101111, 0b01110010, 0b01101100, 0b01100100]
    return "".join(chr(bit) for bit in code)

print("--- Immortal Binary Assertion ---")
original_message = immortal_binary_hello()
print(f"Binary Encoded Message (decoded): {original_message}")
print(f"Data Retained: The string 'Hello World' is directly recoverable from the binary.")

# --- 2. The Intermediate Layer: Soliton Packet Wrapper ---
# Purpose: Encapsulates data in a stabilizing palindromic symmetry to ensure integrity
#          and coherence during transport, particularly across a 64-bit bus.
# Data Retention: The original message is embedded within the packet, flanked by
#          its mirrored counterpart, making the original message directly extractable.

def soliton_packet(message):
    # Enclose data in a stabilizing symmetry
    packet = f"SOLITON_START_{message}_MIRROR_{message[::-1]}_SOLITON_END"
    return packet

print("\n--- Soliton Packet Wrapper ---")
packet_message = soliton_packet(original_message)
print(f"Original Message (from Binary Layer): '{original_message}'")
print(f"Soliton Packet: {packet_message}")
print(f"Data Retained: The original message '{original_message}' is explicitly present and extractable from the packet.")

# --- 3. The Advanced Layer: Fractal Modulation Logic ---
# Purpose: Maps the data string to complex plane coordinates, where it modulates
#          a fractal boundary, embedding the information into spatial logic.
# Data Retention: The original message's characters determine the initial conditions
#          and trajectory of the fractal path. The sequence of complex numbers
#          uniquely represents the input string, allowing for its reconstruction
#          (with a known mapping).

def fractal_modulate(message, iterations=10):
    # Mapping ASCII to coordinate offsets in the fractal set
    # Note: For simplicity, this example uses a direct character-to-float mapping.
    # A robust system would require a predefined, reversible mapping.
    mapping = {char: i * 0.01 for i, char in enumerate(message)}
    # Ensure mapping covers all chars in message if not already covered
    current_max_val = max(mapping.values()) if mapping else 0
    for char_code in sorted(list(set(message) - set(mapping.keys()))):
        current_max_val += 0.01
        mapping[char_code] = current_max_val

    fractal_path = []

    for char in message:
        # Complex coordinate generation based on character value
        # Using a fixed imaginary part for demonstration
        z = complex(mapping[char], 0.1)
        # Recursive iteration to find stable carrier frequency
        for _ in range(iterations):
            # Mandelbrot-like iteration (z = z^2 + c)
            z = z**2 + complex(-0.75, 0.1) # 'c' is a constant for the fractal set
        fractal_path.append(z)

    return fractal_path, mapping

# Function to decode the fractal coordinates back to a string
def decode_fractal_message(complex_path, char_mapping, iterations=10, tolerance=1e-9):
    # Reconstruct the fractal transformation for each unique character in the mapping
    simulated_final_zs = {}
    for char, initial_real_part in char_mapping.items():
        z_sim = complex(initial_real_part, 0.1)
        for _ in range(iterations):
            z_sim = z_sim**2 + complex(-0.75, 0.1)
        simulated_final_zs[char] = z_sim

    decoded_message_chars = []
    for z_in_path in complex_path:
        found_char = None
        for char, simulated_z in simulated_final_zs.items():
            # Compare complex numbers with a tolerance due to floating point arithmetic
            if abs(z_in_path - simulated_z) < tolerance:
                found_char = char
                break
        decoded_message_chars.append(found_char)

    return "".join(char for char in decoded_message_chars if char is not None)


print("\n--- Fractal Modulation Logic ---")
complex_path, char_mapping = fractal_modulate(original_message)
print(f"Original Message (from Binary Layer): '{original_message}'")
print(f"Fractal Path (first 5 coordinates): {[str(c) for c in complex_path[:5]]}...")
print(f"Character Mapping Sample: {{{', '.join(f"'{k}': {v:.2f}" for k, v in list(char_mapping.items())[:5])}}}...")
print(f"Data Retained: The sequence of complex numbers, combined with the character mapping,")
print(f"               allows for the reconstruction of the original message. Each character's unique")
print(f"               value influences the fractal's trajectory, thus encoding the data.")

print("\n--- Fractal Decoding Demonstration ---")
decoded_message = decode_fractal_message(complex_path, char_mapping)
print(f"Decoded Message from Fractal Coordinates: '{decoded_message}'")
print(f"Decoding successful: {decoded_message == original_message}")
```

#### 3. The Advanced Layer (Fractal Modulation)

This is the operational level. The string is mapped to complex plane coordinates $c = a + bi$. As the function iterates, the "Hello World" signal modulates the fractal boundary, becoming a structural part of the coordinate set.

```
# Fractal Mapping Logic
def fractal_modulate(message, iterations=10):
    # Mapping ASCII to coordinate offsets in the fractal set
    mapping = {char: i * 0.01 for i, char in enumerate(message)}
    fractal_path = []

    for char in message:
        # Complex coordinate generation based on character value
        z = complex(mapping[char], 0.1)
        # Recursive iteration to find stable carrier frequency
        for _ in range(iterations):
            z = z**2 + complex(-0.75, 0.1)
        fractal_path.append(z)

    return fractal_path

print(fractal_modulate("Hello World"))
```
